```python
import os
import random
import time

import numpy as np
import pygame as pg
import pymunk
import skimage.measure as measure

import matplotlib.pyplot as plt
import seaborn as sns

from objects import Person, Wall, Train
from pathfinding import Node, Queue, Pathfinder
from simulator import CovidSim
```

```python
heatmap_tensor = np.load("heatmaps/heatmap_tensor.npy")
```

```python
mensa_index = 3
flip_heatmap = np.fliplr(heatmap_tensor[mensa_index])
rotate_heatmap = np.rot90(flip_heatmap, k=1)

plt.imshow(rotate_heatmap, cmap='Spectral', label='Distance to Target Node')
plt.colorbar(label='Distance to Target Node', orientation = 'horizontal',  pad=0.1, aspect=40)
plt.text(710, 458, "4", fontsize=10, weight=500, color="black", ha="center", va='center')
plt.title('Heatmap to building 4 (Mensa)')
plt.xticks(np.arange(0, 800, step=200))
plt.yticks(np.arange(0, 800, step=200))
# plt.savefig("plots/mensa_heatmap.pdf", format="pdf", dpi=400)
plt.show()
```

```python
x, y = np.meshgrid(np.linspace(0, 800, 800), np.linspace(0, 800, 800))

# Normalize the gradient
dx, dy = np.gradient(rotate_heatmap)
norm = np.sqrt(dx**2 + dy**2)
dx, dy = dx/norm, dy/norm

# Create the vector field visualization with a subsampled number of vectors
plt.quiver(x[::20, ::20], y[::20, ::20], -dx[::20, ::20], -dy[::20, ::20], color='r', scale=60) # 50 80 120
plt.text(710, 458, "4", fontsize=10, weight=500, color="black", ha="center", va='center')
plt.title('Vector Field to building 4 (Mensa)')
plt.gca().set_aspect('equal', adjustable='box') # make the plot a square
plt.gca().invert_yaxis()
plt.xticks(np.arange(0, 800, step=200))
plt.yticks(np.arange(0, 800, step=200))
# plt.savefig("plots/mensa_vector_field.pdf", format="pdf", dpi=800)
plt.show()
```
